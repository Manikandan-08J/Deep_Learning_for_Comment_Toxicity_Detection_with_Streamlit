```python

np.random.seed(42)

mu = 50
sigma = 10
n = 1000

pop_a = np.random.normal(mu, sigma, n)
pop_b = np.random.normal(mu, sigma, n)

plt.hist(pop_a, bins=30, alpha=0.5, label = "Population A")
plt.hist(pop_b, bins=30, alpha=0.5, label = "Population B")

plt.title("Two Population Under Null Hypothesis")
plt.legend()
plt.show()
```

```python
sample_a = np.random.choice(pop_a, 30)
sample_b = np.random.choice(pop_b, 30)

mean_dif = sample_a.mean() - sample_b.mean()

print("observed mean difference", mean_dif)
```

```python
num_simulations = 10000
diffs = []

for i in range (num_simulations):
    a=np.random.choice(pop_a, 30)
    b=np.random.choice(pop_b, 30)
    diffs.append(a.mean() - b.mean())

diffs = np.array(diffs)
```

```python
plt.hist(diffs, bins=40, color='skyblue', edgecolor = 'black')
plt.show()
```

```python
p_value = np.mean(np.abs(diffs)>=np.abs(mean_dif))

print(p_value)
```

```python
k = pd.DataFrame({
    'Salary' : np.random.normal(50000, 8000, 100)
})

k
```

```python
sns.boxplot(x=k['Salary'])
plt.title("Outlier Detection Using the BOXPLOT")
plt.show()
```

```python
sns.scatterplot(k['Salary'])
plt.show()
```

```python
from scipy import stats
```

```python
z = np.abs(stats.zscore(k['Salary']))

k[z>1]
```

```python
q1 = k['Salary'].quantile(0.25)
q3 = k['Salary'].quantile(0.75)

iqr = q1 - q3

iqr
```

```python
lower_bound = q1 -1.5*iqr
upper_bound = q3 +1.5*iqr

out = k[(k['Salary'] < lower_bound) | (k['Salary'] > upper_bound)]

out
```

```python
k['Salary'] = np.where(k['Salary'] > upper_bound, upper_bound, np.where (k['Salary'] < lower_bound, lower_bound, k['Salary']))
```

```python
sns.boxenplot(x=k['Salary'])
plt.title("Outliers Detection Using BOXPLOT")
plt.show()
```

```python
k['Salary_log1p'] = np.log1p(k['Salary'])
sns.boxenplot(x=k['Salary_log1p'])
plt.title("Outliers Detection Using BOXPLOT")
plt.show()
```
